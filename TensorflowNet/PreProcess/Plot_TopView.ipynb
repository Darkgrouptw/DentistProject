Helper Function
---

```python
# 畫圖的 Function
def PlotFunction(DataImgPath):
    # 設定大小
    plt.figure(figsize=(15, 30))
    for i in range(len(DataImgPath)):
        img = cv2.imread(DataImgPath[i])
    #     hist, bin = np.histogram(img, bins=np.arange(256))
    
        # 攤平
        flatData = img.reshape([-1])
    
        plt.subplot(len(DataImgPath), 1, (i + 1))
        plt.hist(flatData, bins=256)
        plt.title(DataImgPath[i])
        
        mean = np.mean(flatData)
        std = np.std(flatData)
        print("Mean: ", mean, "Std: ", std)
    plt.show()
```

```python
PlotFunction(Data[:5])
```
